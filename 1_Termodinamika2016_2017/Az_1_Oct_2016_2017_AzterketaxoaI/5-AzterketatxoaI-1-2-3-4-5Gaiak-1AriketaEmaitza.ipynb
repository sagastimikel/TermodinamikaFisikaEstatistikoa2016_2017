```python
# Reference: http://nbviewer.ipython.org/github/demotu/BMC/blob/master/notebooks/CurveFitting.ipynb

def Ajuste(x, y, yerr, conf):
    # 
    yerr      errores en la variable dependiente
    # Estructura para controlar que los errores sean nulos, diberger√≠an los pesos
    w1 = None if (yerr is None or np.sum(yerr)==0) else 1/yerr
    p, cov = np.polyfit(x, y, 1, w=w1, cov=True)       # coefficients and covariance matrix
    yfit = np.polyval(p, x)                            # evaluate the polynomial at x
    perr = np.sqrt(np.diag(cov))                       # standard-deviation estimates for each coefficient
    R2 = np.corrcoef(x, y)[0, 1]**2                    # coefficient of determination between x and y
    resid = y - yfit
    chi2red = np.sum((resid/yerr)**2)/(y.size - 2)
    
    # Confidence intervals predictions
    
    import scipy.stats as stats
    s_err = np.sqrt(np.sum(resid**2)/(y.size - 2))  # standard deviation of the error (residuals)
    x2 = np.linspace(np.min(x), np.max(x), 100)
    y2 = np.linspace(np.min(yfit), np.max(yfit), 100)
    # Confidence interval for the linear fit:
    t = stats.t.ppf(conf, y.size - 2)
    ci = t * s_err * np.sqrt(1./y.size + (x2 - np.mean(x))**2/np.sum((x-np.mean(x))**2))
    # Prediction interval for the linear fit:
    pi = t * s_err * np.sqrt(1. + 1./y.size + (x2 - np.mean(x))**2/np.sum((x-np.mean(x))**2))

    return round(p[1],6), round(p[0],6), round(R2,6)#, figAjuste
```

## 3. Ariketa, ebazpena

>*van der Waals*-en egoera-ekuazioari segitzen dion jariakinaren kasuan,  
>lortu prozesu adiabatiko itzulgarrien adierazpena $T/V$ diagraman.

-----------------------

*van der Waals*-en egoera-ekuazioa:  

$$\left(p + \dfrac{a\,n^{2}}{V^{2}}\right)\left(V-nb\right) = n R\,T$$  

$$\left(p + \dfrac{a}{v^{2}}\right)\left(v-b\right) = R\,T $$  

Prozesu adiabatikoen adierazpena lortzeko bi tresna dugu:  
1. lehenenengo printzipioaren adierazpen diferentziala: $\delta Q_{IG} = dU-\delta W$ .  
2. entropiaren eragiketa-definizioa: $dS = \dfrac{\delta Q_{IG}}{T} \Rightarrow TdS = \delta Q_{IG}$ .  

*van der Waals*-en jariakina sistema hidrostatikoa da, beraz, $\delta W = -p\,dV$ dugu.  

Bestetik, eskatzen digute $T/V$ aldagiak erabiltzea, idatz dezagun bada, $U=U(T,V)$ eta horren adierazpen diferentziala, koefiziente esperimentalen funtzioan:  

$$dU = \left(\dfrac{\partial U}{\partial T}\right)_{V} dT + \left(\dfrac{\partial U}{\partial V}\right)_{T} dV $$  

Sistema hidrostatikoetan, orokorra da honako adierazpen hau: $C_{V} = \left(\dfrac{\partial U}{\partial T}\right)_{V}$ .  

Beraz, adierazpen hau idatz daiteke:  

$$ dU = C_{V} dT + \left(\dfrac{\partial U}{\partial V}\right)_{T} dV $$  

Lehenengo printzipioaren gaian honako adierazpen hau ondorioztatu genuen:  

$$ \left(\dfrac{\partial U}{\partial V}\right)_{T} = \left(\dfrac{C_{p}-C_{V}}{V\alpha}\right)- p$$ .  

Bil dezagun dena adierapen bakar batean:  

$$ T\,dS = C_{V}\,dT + \left[\left(\dfrac{C_{p}-C_{V}}{V\alpha}\right)- p\right]dV + p\,dV$$  

$$ T\,dS = C_{V}\,dT + \left[\left(\dfrac{C_{p}-C_{V}}{V\alpha}\right)- p + p\right]\,dV$$  

$$ T\,dS = C_{V}\,dT + \left[\left(\dfrac{C_{p}-C_{V}}{V\alpha}\right)\right]\,dV$$  

Eta prozesua adiabatiko itzulgarria denez, $T\,dS = 0$. Honako hau da, hortaz, integratu beharreko ekuazio diferentziala:  

$$ 0 = C_{V}\,dT + \left[\left(\dfrac{C_{p}-C_{V}}{V\alpha}\right)\right]\,dV$$.  

Bestetik, ekar dezagun gogora *Mayer*-en erlazioa: $C_{p} - C_{V} = \dfrac{T\,V\,\alpha^{2}}{\kappa_{T}}$.  
Eta $\alpha$ eta $\kappa_{T}$ koefiziente esperimentalen definizioak kontuan hartuz, honako era honetan idatz daiteke aipatutako erlazioa:  

$$C_{p} - C_{V} = \dfrac{T\,V\,\alpha^{2}}{\kappa_{T}}\Rightarrow  T \left(\dfrac{\partial p}{\partial T}\right)_{V}$$.  

$$ 0 = C_{V}\,dT - T \left(\dfrac{\partial p}{\partial T}\right)_{V}\,dV \Rightarrow \boxed{-\dfrac{C_{V}}{T}\,dT = \left(\dfrac{\partial p}{\partial T}\right)_{V}\,dV}$$.  

Kalkula dezagun egoera-ekuazioaren deribatu partziala: $\left(\dfrac{\partial p}{\partial T}\right)_{V}$ .  

$$ p = \dfrac{RT}{(V-b)} -\dfrac{a^{2}}{V^{2}} \Rightarrow \left(\dfrac{\partial p}{\partial T}\right)_{V} = \dfrac{R}{(V-b)}$$.  

Eta goragoko adierazpenean ordezkatuz, honakoa da integratu beharreko ekuazio diferentziala:  

$$ \boxed{\dfrac{C_{V}}{T}\,dT = \dfrac{R}{(V-b)}\,dV} $$

## 4. Ariketa, ebazpena


>*van der Wals*-en egoera-ekuazioari segitzen dion jariakinaren kasuan,  
>lortu zer forma duen barne-energiak, $(T, V)$ aldagaiak independentetzat hartuta.

-------------------

Aurreko ariketan, har dezagun barne-energiaren adierazpen diferentziala, nahiz eta $T/V$ aldagietan idatzita dagoen:

$$ dU = C_{V} dT + \left[T \left(\dfrac{\partial p}{\partial T}\right)_{V}- p \right]dV $$

$$ dU = C_{V} dT + \left[T \dfrac{R}{(V-b)}- \left\{ \dfrac{RT}{(V-b)} -\dfrac{a^{2}}{V^{2}}\right\} \right]dV $$

$$ dU = C_{V} dT + \left[ - \left\{  -\dfrac{a^{2}}{V^{2}}\right\} \right]dV $$

Eta, integratuz gero, honako hau lortzen da:

$$ \boxed{\Delta U = C_{V}\Delta T - a^{2}\,\left(\dfrac{1}{V_{f}}-\dfrac{1}{V_{i}}\right)}$$.
